(e) Plot by category

<font size = "3">

- Read "data/wdi_urban.csv"<br>
This contains data from 1980 and 2020

- Create a list with the unique values of "year"
- Create a plot with a ...
    - Histogram of "prop_urbanpopulation" <br>
    for each year
    - Set the histogram option ``` alpha = 0.5 ``` for transparency
    - All the histograms should be overlapping
    - Create a legend with the list of years
    - Label the axes and the title

What do we learn from the graph?

HINT: See the example at the end of Lecture 10


````python
# Write your own code here
wdi_urban = pd.read_csv("data/wdi_urban.csv")
list_unique_year = pd.unique(wdi_urban["year"])

for category in list_unique_year:
    df   = wdi_urban.query("year == @category")
    plt.hist(x = df["prop_urbanpopulation"], alpha = 0.5)
plt.xlabel("Proportion of the Population is Urban")
plt.ylabel("Number of Countries")
plt.title("Number of Countries vs Proportion of the Population is Urban")
plt.legend(labels = list_unique_year,
title = "Years")
plt.show()

# From this graph, we learn that as the years progressed, the proportion of the population that is urban increases. Indicated by the orange histograms(2020) being more towards the right end than the blue histograms(1980)
````
